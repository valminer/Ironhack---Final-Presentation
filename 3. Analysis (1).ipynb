# 3.2. Analysis


In this section, I will analyze the clusters by looking at their features. Part of the feature analysis was run on the basis of the Spotify descriptions, part was instead based on assumptions. 
In particular, in the case of valence, by exploring the dataset I reached the (subjective) conclusion that only songs with a valence above .80 could actually be considered as "cheerful". Between .34 and .8 the level of valence decreases to a more angry, energetic style, typical for example of trap or rap music. Below 0.34 I could instead find many "sad" or "romantic" songs. In order to spot this, I first divided the data between "happy" and "sad/angry", and then I calculated the average to capture the general mood of the playlist.


I could then reach the final conclusion that cluster 0 is composed by "darker" and more "down" music, very close to urban, and cluster 1 is instead represented by more sad, lower energy songs.

__RECALL__:

__Danceability__ = A value of 0.0 is least danceable and 1.0 is most danceable.

__Energy__ = Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale.

__Acousticness__ = A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic. 

__Instrumentalness__= The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0. 

__Liveness__= Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live. 

__Tempo__=  In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.

__Valence__=Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

```python
cluster_zero = df_explore[df_explore["cluster"] == 0] 
```

```python
cluster_zero
```

```python
cluster_zero['Streams'].sum()
```

```python
danceability = cluster_zero['danceability'].apply(lambda x: 'Danceable' if x > 0.4 else 'Not Danceable')
danceability.value_counts() #majority of danceable songs 
```

```python
energy = cluster_zero['energy'].apply(lambda x: 'High Energy' if x > 0.5 else 'Low Energy')
energy.value_counts() #majority of high energy songs
```

```python
acousticness = cluster_zero['acousticness'].apply(lambda x: 'Acoustic' if x > 0.8 else 'Not Acoustic')
acousticness.value_counts() #high confidence that there are only not acoustic tracks, i.e. there are many sounds that are electric 
```

```python
valence = cluster_zero['valence'].apply(lambda x: 'Happy' if x > 0.8 else 'Sad/Angry')
valence.value_counts()
```

```python
cluster_zero['valence'].mean()
```

```python
tempo = cluster_zero['tempo'].apply(lambda x: 'high tempo' if x > 130 else 'lower tempo')
tempo.value_counts() 
```
